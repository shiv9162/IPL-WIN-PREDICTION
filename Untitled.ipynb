```python
import numpy as np
import pandas as pd
```

```python
#add data set
match = pd.read_csv('matches.csv')
delivery = pd.read_csv('deliveries.csv')
```

```python
#total score in the first inning and second inning and reset index convert to data frame

total_score_df =delivery.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
```

```python
#total score in first inning in all the match
total_score_df =total_score_df[total_score_df['inning'] == 1]
```

```python
#SHOW THE DATA
total_score_df
```

```python
#merge of data of total_score_df to merge to match.csv
match_df = match.merge(total_score_df[['match_id','total_runs']],left_on='id',right_on='match_id')
```

```python
#show the match data set
match_df
```

```python
#list of team
match_df['team1'].unique()
```

```python
#team list
teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals'
]
```

```python
#change the name of team
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
#after change team name then add the  
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
#after remove the team remaining team  play
match_df.shape
```

```python
#without di applied
match_df = match_df[match_df['dl_applied'] == 0]
```

```python
#we have to find math_id,city,winner and total_runs in match data
match_df = match_df[['match_id','city','winner','total_runs']]
```

```python
#merge the data in delivery data of match id
delivery_df = match_df.merge(delivery,on='match_id')
```

```python
# how many ball are in second inning
delivery_df = delivery_df[delivery_df['inning'] == 2]
```

```python
#after merge the data in delivery
delivery_df
```

```python
#add the run at how many balls have been thrown 
delivery_df['current_score'] = delivery_df.groupby('match_id').cumsum()['total_runs_y']
```

```python
# how many runs are remainder
delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score']
```

```python
#how many ball are remainder
delivery_df['balls_left'] = 126 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
#show the data
delivery_df
```

```python
#how many wicket are left
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")#nan value replace from 0
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")#if x value is same then same(o) otherwise(1) 
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')#player_dismissed colum convert in int
wickets = delivery_df.groupby('match_id').cumsum()['player_dismissed'].values#how many wicket are go on
delivery_df['wickets'] = 10 - wickets #how many wicket are left
delivery_df.head()#show the data
```

```python
#Show the data
delivery_df.head()
```

```python
# crr = runs/overs
delivery_df['crr'] =(delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
```

```python
#require run rate rrr=runs_left*6/balls_left
delivery_df['rrr'] =(delivery_df['runs_left']*6)/delivery_df['balls_left']
```

```python
#winning the match indicate 1 other wise 0
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
#lost the match
delivery_df['result'] = delivery_df.apply(result,axis=1)
```

```python
#require formate are find out
final_df=delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr','result']]
```

```python
#result was suffer because of ek match ke sari ball ek sath hai that reason bias are create so result are suffer
final_df = final_df.sample(final_df.shape[0])
```

```python
#show the sample
final_df.sample()
```

```python
#missing the value of and drop it final_df.isnull().sum()
final_df.dropna(inplace=True)
```

```python
# when ball_left is 0 then except the value otherwise accepted
final_df = final_df[final_df['balls_left'] != 0]
```

```python
#model building
X = final_df.iloc[:,:-1] #all row except last column
y = final_df.iloc[:,-1]# all row and last column
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
#Show the data
X_train
```

```python
#implement column transformer & onehot encoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
#implement logistic regression,random forest classifier 
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
#pipeline distributed two step 1.column trf 2.logisticregression
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
#model was trained
pipe.fit(X_train,y_train)
```

```python
#y are predict
y_pred = pipe.predict(X_test)
```

```python
#logistic regression will predicted
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
#Who won after throwing some balls
pipe.predict_proba(X_test)[10]
```

```python
#match summmary row
def match_summary(row):
    print("Batting Team-" + row['batting_team'] + " | Bowling Team-" + row['bowling_team'] + " | Target- " + str(row['total_runs_x']))
    
```

```python
#Who will win by giving complete match data to whom?
def match_progression(x_df,match_id,pipe):
    match = x_df[x_df['match_id'] == match_id]
    match = match[(match['ball'] == 6)]
    temp_df = match[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr']].dropna()
    temp_df = temp_df[temp_df['balls_left'] != 0]
    result = pipe.predict_proba(temp_df)
    temp_df['lose'] = np.round(result.T[0]*100,1)
    temp_df['win'] = np.round(result.T[1]*100,1)
    temp_df['end_of_over'] = range(1,temp_df.shape[0]+1)
    
    target = temp_df['total_runs_x'].values[0]
    runs = list(temp_df['runs_left'].values)
    new_runs = runs[:]
    runs.insert(0,target)
    temp_df['runs_after_over'] = np.array(runs)[:-1] - np.array(new_runs)
    wickets = list(temp_df['wickets'].values)
    new_wickets = wickets[:]
    new_wickets.insert(0,10)
    wickets.append(0)
    w = np.array(wickets)
    nw = np.array(new_wickets)
    temp_df['wickets_in_over'] = (nw - w)[0:temp_df.shape[0]]
    
    print("Target-",target)
    temp_df = temp_df[['end_of_over','runs_after_over','wickets_in_over','lose','win']]
    return temp_df,target
    
```

```python
#Predicating after sending the complete data of the 74 match
temp_df,target = match_progression(delivery_df,74,pipe)
temp_df
```

```python
#all summary of match(74)
import matplotlib.pyplot as plt
plt.figure(figsize=(18,8))
plt.plot(temp_df['end_of_over'],temp_df['wickets_in_over'],color='yellow',linewidth=3)
plt.plot(temp_df['end_of_over'],temp_df['win'],color='#00a65a',linewidth=4)
plt.plot(temp_df['end_of_over'],temp_df['lose'],color='red',linewidth=4)
plt.bar(temp_df['end_of_over'],temp_df['runs_after_over'])
plt.title('Target-' + str(target))
```
